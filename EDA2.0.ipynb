```python
threshold = 0.55  # You can adjust this threshold as needed

# Find indices where correlation is greater than or equal to the threshold
strong_positive_corr = (corr_matrix > threshold) & (corr_matrix < 1.0)

# Find indices where correlation is less than or equal to the negative threshold
strong_negative_corr = (corr_matrix < -threshold) & (corr_matrix > -1.0)

# Use boolean indexing to extract the values with strong positive or negative correlation
strong_positive_values = corr_matrix[strong_positive_corr].stack()
strong_negative_values = corr_matrix[strong_negative_corr].stack()

# Print the results
print("Strong positive correlations:")
print(strong_positive_values)

print("\nStrong negative correlations:")
print(strong_negative_values)
```
